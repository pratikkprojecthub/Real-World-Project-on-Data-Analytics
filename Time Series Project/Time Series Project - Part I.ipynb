```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

### To avoid warnings
from warnings import filterwarnings
filterwarnings('ignore')
```

# Stocks of Tech Giants

```python
path=r'C:\Users\Asus\Desktop\python101\individual_stocks_5yr'
list_of_company = ['AMZN_data.csv', 'MSFT_data.csv', 'AAPL_data.csv','GOOG_data.csv']

#blank dataframe
all_data = pd.DataFrame()

for file in list_of_company:
    current_df = pd.read_csv(path+"/"+file)
    all_data = pd.concat([all_data, current_df])
    
all_data.shape
```

# Analyzing Closing Price of all the Stocks 

```python
giant = all_data['Name'].unique()
```

```python
plt.figure(figsize=(18,12))
for i, company in enumerate(giant,1):
    plt.subplot(2, 2, i)
    df=all_data[all_data['Name']==company]
    plt.plot(df['date'],df['close'])
    plt.title(company)
```
